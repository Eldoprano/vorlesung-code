# Read file

```python
file = open("random_vectors.txt", "r")
large_str = file.read()
file.close()
```

# Convert strings to tuples

```python
tokens = large_str.split()
tokens[:20]
```

```python
eval( "(3,5)" )
```

```python
points = [eval(t) for t in tokens]
points[:20]
```

# Plot points

```python
x = [p[0] for p in points]
y = [p[1] for p in points]
```

```python
x[:5]
```

```python
y[:5]
```

```python
import matplotlib.pyplot as plt

plt.scatter(x,y, alpha=0.1)

plt.xlim(-15,15)
plt.ylim(-5,15)
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

# Flying with Python

Generating random vectors from a multi-variate normal distribution with Python is much easier!

```python
import numpy as np
```

```python
SIGMA = np.array([[10,0], \
                   [0,1], \
                 ])
M = np.random.multivariate_normal([5,7], SIGMA, size=10000)
```

```python
M
```

```python
import matplotlib.pyplot as plt

plt.scatter(M[:,0], M[:,1], alpha=0.1)

plt.xlim(-15,15)
plt.ylim(-5,15)
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```
