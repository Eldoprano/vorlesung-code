```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000
mu = np.array([0,0])
cov_low = np.array([[10,0],
                    [0,2]])
data_low_cov = np.random.multivariate_normal(mu, cov_low, N)

cov_high = np.array([[10,2],
                     [2,10]])
data_high_cov = np.random.multivariate_normal(mu, cov_high, N)

det_low = np.linalg.det(cov_low)
det_high = np.linalg.det(cov_high)

plt.subplot(1,2, 1)
plt.xlim(-15,15)
plt.ylim(-15,15)
plt.scatter(data_low_cov[:,0], data_low_cov[:,1], marker=".", alpha=0.2)
plt.title(f"Data sampled from MVN\nwith low covariance values\nDet: {det_low:.2f}") 

plt.subplot(1,2, 2)
plt.xlim(-15,15)
plt.ylim(-15,15)
plt.scatter(data_high_cov[:,0], data_high_cov[:,1], marker=".", alpha=0.2)
plt.title(f"Data sampled from MVN\nwith high covariance values\nDet: {det_high:.2f}") 

plt.show()
```
