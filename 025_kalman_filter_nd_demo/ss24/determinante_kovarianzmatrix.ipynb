```python
import numpy

mu = [5.0, 5.0]
C = [[1.0,  0.0],
     [0.0,  10.0]]

D = numpy.random.multivariate_normal(mean=mu, cov=C, size=1000)

import matplotlib.pyplot as plt
plt.scatter(D[:,0], D[:,1])
plt.xlim(-20,+20)
plt.ylim(-10,+30)

determinante = numpy.linalg.det( C )

plt.title(f"Determinante ist hier: {determinante:.2f}")
plt.show()
```
