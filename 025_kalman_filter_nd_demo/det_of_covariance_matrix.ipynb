# Introduction

Here we show that the determinant of the covariance matrix of data points grows, if the data is more dispersed.

```python
import numpy as np

n = 2
nr_points = 1000
SIGMA = np.array([[2.0, 0.0],
                  [ 0.0, 5.0]])
data = np.random.multivariate_normal(np.zeros(n), SIGMA, size=nr_points)
```

```python
data[:5]
```

```python
import matplotlib.pyplot as plt
plt.scatter(data[:,0], data[:,1] )
plt.show()
```

```python
SIGMA_EST = np.cov(data, rowvar=False)
SIGMA_EST
```

```python
np.linalg.det(SIGMA_EST)
```
