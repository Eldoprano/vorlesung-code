Wieviele Stunden Audio hört ein Mensch in etwa in seinem Leben?

# Klassifikation der Einsatzgebiete

Viele Anwendungsbeispiele transformieren eine Eingabe X in eine Ausgabe Y

mit X,Y ist ein Element von {Text, Wort, Bild, Video, Merkmalsvektor, Audiosignal}

| X | Y | Anwendungsbeispiel | Beispiel |
|---|---|---|---|
| Text | Text | Neural Machine Translation (NMT) | [DeepL](https://www.deepl.com/de/translator)<br>[Github Copilot](https://github.com/features/copilot) |
| Text | Wort | Dokumentenklassifikation (Art, Vertraulich/Nicht vertraulich) | [Titus Software](https://www.youtube.com/watch?v=OSdPf-LUovo) |
| Text | Bild | Bildgenerator | [DALL-E 2](https://openai.com/dall-e-2/) | |
| Text | Merkmalsvektor | Sentiment Analysis (Stimmungserkennung), Echtheitserkennung | [Quora Insincere Questions Classification](https://www.kaggle.com/c/quora-insincere-questions-classification) |
| Text | Audiosignal | Text-to-Speech, Voice Cloning | |
| Bild | Bild | Image-to-Image-Translation<br>Skizze eines Bildes --> Komplettes Bild<br>Bild bei Nacht --> Bild bei Tag<br>Bild bei Nebel --> Bild ohne Nebel<br>Bild ohne Farbe --> Bild mit Farbe| [DeOldify](https://www.youtube.com/watch?v=UDC712hJmXo) |
| Bild | Text | Image Captioning<br>Beschreibung des Bildes in Textform | [Image Captioning Demo](https://www.youtube.com/watch?v=RlNjzz358Jc)  |
| Bild | Wort | Bildklassifikation |   |
| Bild | Merkmalsvektor | Objektlokalisation | [YOLO v7](https://www.youtube.com/watch?v=bt3JLYfFaKE) |
| Bild | Audiosignal | Sound-Synthese | |
| Bild | Video | Einzelbild --> Bewegtes Bild |  [Beispiel von NVIDIA](https://developer.nvidia.com/blog/ai-looping-videos-from-still-images/) | 
| Audiosignal | Text | Speech-To-Text | |
| Audiosignal | Wort | Einzelkommandoerkennung | |
| Audiosignal | Merkmalsvektor | Emotion Recognition, Sprechererkennung | |
| Audiosignal | Audiosignal | Voice-Cloning | |
| Audiosignal+Bild | Audiosignal+Bild | Neu-Vertonung von Videos in anderer Sprache | [Synthesia](https://www.youtube.com/watch?v=PhZpnDqNNmY) |
