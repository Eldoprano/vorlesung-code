# Introduction

In this notebook, I try to show that the determinant of a covariance matrix can be interpreted as a measure of the dispersion of the data points.

# Comparison

In the following code, the function `np.random.multivariate_normal` generates data following a multivariate normal distribution, where the first argument is the mean vector and the second argument is the covariance matrix. The `np.cov` function calculates the covariance matrix from the data, and the `np.linalg.det` function calculates the determinant of a matrix.

In the generated plot, you can see how the high covariance data is more dispersed and therefore has a larger determinant than the low covariance data.

```python
import numpy as np
import matplotlib.pyplot as plt

# make sure, everything is reproducible
np.random.seed(0)

# generate two datasets
data_low_cov  = np.random.multivariate_normal([0, 0], [[1, 0], [0, 1]], 1000)
data_high_cov = np.random.multivariate_normal([0, 0], [[10, 5], [5, 10]], 1000)

# compute covariance matrices
cov_low = np.cov(data_low_cov, rowvar=False)
cov_high = np.cov(data_high_cov, rowvar=False)

# compute determinants of the covariance matrices
det_low = np.linalg.det(cov_low)
det_high = np.linalg.det(cov_high)

# plot the data points and show the values of the determinants
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.xlim(-15,15)
plt.ylim(-15,15)
plt.scatter(data_low_cov[:, 0], data_low_cov[:, 1])
plt.title('Low Covariance Data\nDeterminant: {:.2f}'.format(det_low))

plt.subplot(1, 2, 2)
plt.xlim(-15,15)
plt.ylim(-15,15)
plt.scatter(data_high_cov[:, 0], data_high_cov[:, 1])
plt.title('High Covariance Data\nDeterminant: {:.2f}'.format(det_high))

plt.show()

print("cov_low:")
print(cov_low)

print("cov_high:")
print(cov_high)
```
